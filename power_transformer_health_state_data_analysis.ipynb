```python
# Import necessary libraries.
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read data from csv file.
csv_file = './data/Health index2.csv'
df = pd.read_csv(csv_file)
df.head()
```

```python
# Check columns data type.
df.info()
```

```python
# Group by range of data in Health Index column.
df.groupby(pd.cut(df['Health index'], np.arange(0, 100, 10))).count()
```

```python
# Plot group by range of data in Health Index column as bar chart.
group_index_df_a = df.groupby(pd.cut(df['Health index'], np.arange(0, 100, 10))).count()
group_index_df_b = group_index_df_a['Health index']
group_index_df_b.plot(kind='bar', figsize=(15,10))
```

```python
# Plot pairwise relationships in a dataset.
sns.pairplot(data=df)
```
